```python
import matplotlib.pyplot as plt
import numpy as np
aa=np.array([1,2.5,3])
bb=np.array([1,2,3])
%matplotlib inline
plt.plot(aa,bb)
plt.show()
```

```python
plt.plot(aa, bb, 'go', linewidth=2, markersize=12)
```

```python
plt.plot([1, 2, 3], [1, 2, 3], 'go-', label='line 1', linewidth=2)
plt.plot([1, 2, 3], [1, 4, 9], 'r^', label='line 2')
plt.legend()
plt.show()
```

```python
plt.plot([2,3,4],[4,2,6],'go',[3,2,0],[1,5,8],'r^')
plt.title("Vẽ biểu đồ")
plt.xlabel("Trục x")
plt.ylabel("Trục y")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot([1,2,3],[3,1,2],'go')
plt.title("Hình 1")

plt.subplot(1,2,2)
plt.plot([5,5,6],[1,2,4])
plt.title("Hinh 2")
plt.suptitle("Tiêu đề chung")
```

```python
from IPython.core.pylabtools import figsize
x=np.arange(1,5)
y=x**2
fig,ax=plt.subplots(nrows=2,ncols=2,figsize =(4,4))
ax[0,0].plot([3,1],[3,1],)
ax[0,0].set_title("Hình 1")
ax[0,1].plot([1,2,3],[4,3,1],'go')
ax[0,1].set_title("Hình 2")
ax[1,0].plot(x,y,'r^')
ax[1,0].set_title("Hình 3")
ax[1,1].plot([2,3,1],[5,4,0],'b')
ax[1,1].set_title("Hình 4")
```

```python
import numpy as np
import matplotlib.pyplot as plt

lable=["Cột 1","Cột 2","cột 3","Cột 4"]
y=[10,20,30,40]
x=[20,40,60,80]
width=0.30
index =np.arange(4)
plt.bar(index,x,width,color='green',label="2020")
plt.bar(index+width,y,width,color='red',label="2021")
plt.xticks(index,lable)
plt.legend()
plt.show()
```
